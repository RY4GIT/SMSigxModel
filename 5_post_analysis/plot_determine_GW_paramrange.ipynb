```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
input_path = r"..\2_GLUE_prerun\results\LittleWashita-2023-08-29-determine-GW-param-range-NWM\LittleWashita-2023-08-29"
output_path = r".\results\LittleWashita-2023-08-29-determine-GW-param-range-NWM"
if not os.path.exists(output_path):
    os.makedirs(output_path)
```

```python
evaluation_metrics = pd.read_csv(os.path.join(input_path, "evaluation_metrics.csv"))
evaluation_metrics.set_index(keys="run_id", inplace=True)
evaluation_metrics.head()
```

```python
parameters = pd.read_csv(os.path.join(input_path, "prior_parameters.csv"))
parameters.set_index(keys="run_id", inplace=True)
parameters.head()
```

```python
all_results = pd.concat([evaluation_metrics, parameters], axis=1)
all_results.head()
```

```python
target_params = ["max_gw_storage", "Cgw", "expon", "smcmax"]
eval_metrics_names = ["NSE on Flow", "NSE on Soil", "KGE on Flow", "KGE on Soil"]
```

```python
def plot_parameter_dotty(all_results, target_params, eval_metrics, output_path):
    nparas = len(target_params)

    for target_eval in eval_metrics:
        f, axes = plt.subplots(
            nparas, 1, figsize=(7, 7), constrained_layout=True
        )

        for i, target_para in enumerate(target_params):
            ax = axes[i]
            ax.scatter(
                all_results[target_para],
                all_results[target_eval],
                alpha=0.5,
            )
            ax.tick_params(axis="both", which="major", labelsize=16)
            ax.set_xlabel(target_para, fontsize=16)
            ax.set_ylabel(target_eval, fontsize=16)
            ax.set_ylim([-1, 1])

        save_path = os.path.join(output_path, f"param_dotty_{target_eval}.png")
        f.savefig(save_path, dpi=600)
        plt.close(f)
```

```python
plot_parameter_dotty(all_results, target_params, eval_metrics_names, output_path)
```
