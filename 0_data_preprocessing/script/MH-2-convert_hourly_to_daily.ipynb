```python
import pandas as pd
import numpy as np
import os
```

```python
hourly_forcing_path = r"G:\Shared drives\Ryoko and Hilary\SMSigxModel\analysis\data\Mahurangi\forcing_hourly_1998_2001.csv"
hourly_test_path = r"G:\Shared drives\Ryoko and Hilary\SMSigxModel\analysis\data\Mahurangi\test_hourly_1998_2001_sm_basinavg.csv"
```

```python
df_hourly_forcing = pd.read_csv(hourly_forcing_path)
df_hourly_forcing['time'] = pd.to_datetime(df_hourly_forcing['time'])
df_hourly_forcing.set_index('time', inplace=True)
```

```python
df_daily_forcing = df_hourly_forcing.resample('D').sum()
df_daily_forcing.plot()

output_path = r"G:\Shared drives\Ryoko and Hilary\SMSigxModel\analysis\data\Mahurangi\forcing_daily_1998_2001.csv"
df_daily_forcing.to_csv(output_path)
```

```python
df_hourly_test = pd.read_csv(hourly_test_path)
df_hourly_test['Time'] = pd.to_datetime(df_hourly_test['Time'])
df_hourly_test.set_index('Time', inplace=True)
df_hourly_test.head()
```

```python
df_daily_test = pd.DataFrame()
```

```python
df_daily_test['Rainfall'] = df_hourly_test['Rainfall'].resample('D').sum()
df_daily_test['Flow'] = df_hourly_test['Flow'].resample('D').sum()
df_daily_test['Soil Moisture Content'] = df_hourly_test['Soil Moisture Content'].resample('D').mean()
df_daily_test['Direct Runoff'] = df_hourly_test['Direct Runoff'].resample('D').sum()
df_daily_test['GIUH Runoff'] = df_hourly_test['GIUH Runoff'].resample('D').sum()
df_daily_test['Lateral Flow'] = df_hourly_test['Lateral Flow'].resample('D').sum()
df_daily_test['Base Flow'] = df_hourly_test['Base Flow'].resample('D').sum()
df_daily_test['Total Discharge'] = df_hourly_test['Total Discharge'].resample('D').sum()
```

```python
df_daily_test['Timestep'] = range(len(df_daily_test))
df_daily_test['Timestep']
```
