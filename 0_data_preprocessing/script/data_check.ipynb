```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
fn = r"G:\Shared drives\Ryoko and Hilary\SMSigxModel\analysis\2_data_input\LittleWashita\full\test_sm_basinavg.csv"
df = pd.read_csv(fn)
df.head()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df.set_index(['Time'], inplace=True)
```

```python
fig, ax1 = plt.subplots(figsize=(15,2.5))

var_to_compare_with = "Soil Moisture Content"
var_to_compare_with2 = "precip_rate"
# var_to_compare_with2 = "Flow"

# plot the flow data on the first y-axis
df[var_to_compare_with].plot(ax=ax1, color='blue', alpha=0.5)
ax1.set_ylabel(var_to_compare_with, color='blue')
ax1.tick_params('y', colors='blue')

# create a second y-axis object
ax2 = ax1.twinx()

# plot the precipitation data on the second y-axis
df[var_to_compare_with2].plot(ax=ax2, color='orange', alpha=0.5)
ax2.set_ylabel(var_to_compare_with2, color='orange')
ax2.tick_params('y', colors='orange')

# set x-axis label and title
ax2.set_xlabel('Month')
ax1.set_title('Data comarison')

ax1.set_xlim([datetime.datetime(2011, 5, 1), datetime.datetime(2011, 6, 1)])
# show the plot
plt.show()
```
