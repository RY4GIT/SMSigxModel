## Read data

```python
filename = r"G:\Shared drives\Ryoko and Hilary\SMSigxModel\analysis\1_pre_calibration\results\Coweeta-2023-11-22-KGE-Flow\DDS_allresults.csv"
df = pd.read_csv(filename)
print(len(df))
df.head()
```

```python
# parameter_names = ["parbb","parsatdk","parslop","parsmcmax","parwltsmc","parmax_gw_storage"]
parameter_names = ["parbb","parsatdk","parslop","parsmcmax","parwltsmc","parmax_gw_storage","parCgw","parexpon","parK_nash","parrefkdt","partrigger_z_fact","parK_lf"] #,"parnum_nash_storage"]
# parameter_names = ["parbb","parsatdk","parsatpsi","parslop","parsmcmax","parwltsmc","parmax_gw_storage","parCgw","parexpon","parK_nash","parrefkdt","partrigger_z_fact","paralpha_fc","parK_lf","parD","parnum_nash_storage"]
```

## Get dotty plots on the parameter values

```python
num_plots = len(parameter_names)
fig, axes = plt.subplots(4, 4, figsize=(15, 12))  # Adjust figsize as needed
for i, parameter_name in enumerate(parameter_names):
    row, col = divmod(i, 4)
    axes[row, col].scatter(df[parameter_name].values, df["like1"].values, alpha=0.5)
    axes[row, col].set_xlabel(parameter_name)
    axes[row, col].set_ylabel("KGE")
    axes[row, col].set_ylim([-0.41, 1])
fig.tight_layout()
```
