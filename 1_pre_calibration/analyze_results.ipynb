```python
df = pd.read_csv(filename)
df.head()
```

```python
parameter_names = ["parbb","parsatdk","parsatpsi","parslop","parsmcmax","parwltsmc","parmax_gw_storage","parCgw","parexpon","parK_nash","parrefkdt","partrigger_z_fact","paralpha_fc","parK_lf","parnum_nash_storage"]
# parameter_names = ["parbb","parsatdk","parsatpsi","parslop","parsmcmax","parwltsmc","parmax_gw_storage","parCgw","parexpon","parK_nash","parrefkdt","partrigger_z_fact","paralpha_fc","parK_lf","parD","parnum_nash_storage"]
```

```python
num_plots = len(parameter_names)
fig, axes = plt.subplots(4, 4, figsize=(15, 12))  # Adjust figsize as needed
for i, parameter_name in enumerate(parameter_names):
    row, col = divmod(i, 4)
    axes[row, col].scatter(df[parameter_name].values, df["like1"].values, alpha=0.5)
    axes[row, col].set_xlabel(parameter_name)
    axes[row, col].set_ylabel("KGE")
    axes[row, col].set_ylim([-1, 1])
fig.tight_layout()
```
